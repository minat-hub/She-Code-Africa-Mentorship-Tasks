```
#Plotting San Juan unemployment rate around the years surrounding 1990s Recession
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(30,8))
group = c.loc[c['Year']==1990]
sns.pointplot(group['Month'], group['Rate'], alpha=0.8, ax=ax1)
gp = c.loc[c['Year']==1991]
sns.pointplot(gp['Month'], gp['Rate'], alpha=0.8, ax=ax2)
ax2.set_ylabel("")
gop = c.loc[c['Year']==1992]
sns.pointplot(gop['Month'], gop['Rate'], alpha=0.8, ax=ax3)
ax3.set_ylabel("")
plt.show()

fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(30,8))
group_2 = c.loc[c['Year']==1993]
sns.pointplot(group_2['Month'], group_2['Rate'], alpha=0.8, ax=ax1)
gp_2 = c.loc[c['Year']==1994]
sns.pointplot(gp_2['Month'], gp_2['Rate'], alpha=0.8, ax=ax2)
ax2.set_ylabel("")
gop_2 = c.loc[c['Year']==1995]
sns.pointplot(gop_2['Month'], gop_2['Rate'], alpha=0.8, ax=ax3)
ax3.set_ylabel("")
plt.show()

```

In the graph above we will be looking at how unemployment rose and fell around the 1990s Recession month by month. In 1990 the month of Jan started with unemployment rate of 15.0 and ended with the month of december with 12.5, The recession started July of 1990 and ended March 1991. The Unemployment rate started increasing exponentially after March of 1991. Impact was felt from August of 1991 and hit an all time high in Jan 1992. Seems the effect of Recession is not felt in unemployment rate until months after. Another thing we can observe is that Jan has a consistent rate of unemployment and this supports our argument that this might be because companies lay off workers in the 4th quarter of the year.

Let us study the graph above. In 1992, the unemployment rate hits an all time high after the gulf war recession that lasted for a period of 8months, This recession was caused by Iraq invadeing Kuwait. This resulted in a spike in the price of oil in 1990, which caused manufacturing trade sales to decline. This was combined with the impact of manufacturing moving offshore as the provisions of the North American Free Trade Agreement (NAFTA) kicked in. In addition, the leveraged buyout of United Airlines triggered a stock market crash. All these contributed to many layoffs and a hiring freeze because in 1992 GDP growth and Job creation were weak in the united states(Approximately 1.621 million jobs were shed during the recession).

In 1995, We can see that the unemployment rate was lower than in the other years and this was because in the spring of 1994 GDP growth surged and the number of jobs created (3.85 million) set a record that has yet to be surpassed as of 2015. But 1995 would bring a pause in economic growth, primarily because the Federal Reserve raised interest rates from 3% to 6% beginning in late 1994 to prevent inflation from rising after such rapid growth along with two government shutdowns that slowed the economy. The pause was short-lived, however, as the economy adjusted and the surge of investment in the Dot-Com bubble would jumpstart the economy beginning in late 1995. 1996 saw a return to steady growth, and in May 1997 unemployment rose slightly but fell in 1998 because for the first time in years, Usa has Surplus of 69.2 Billion dollars unlike the year before it that they had deficits. The prosperity of the us economy, combined with the Omnibus Budget Reconciliation Act of 1990 and Omnibus Budget Reconciliation Act of 1993 (which raised taxes and restrained spending), allowed the federal government to go from a 290 billion dollars deficit in 1992 to a record 236.4 billion dollars surplus in 2000. The reduction in government borrowing freed up capital in markets for businesses and consumers, causing interest rates on loans to fall creating a cycle that only reinforced growth. Government debt increased from 5.02 trillion dollars in 1990 to 5.413 in 1997 and flatlined, barely increasing to $5.674 in 2000.

2001 saw a rise in unemployment because of the 9/11 recession(March 2001–November 2001) which was caused by The collapse of the dotcom bubble, the 9/11 attacks, and a series of accounting scandals at major U.S. corporations contributed to this relatively mild contraction of the U.S. economy.
