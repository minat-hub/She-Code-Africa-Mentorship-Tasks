Examinations in Nigerian schools dated back to the advent of formal education in the country in the 1800s and it was patterned after the British system. As such, the 1987 Ordinance made provision for examinations in schools that have attained the requisite percentage of proficiency. Towards this end, all secondary schools in the country are expected to teach their subjects to meet the requirements of examination bodies for the senior secondary school certificate organized by the West African Examinations Council and the National Examinations Commission.

A large number of student registered for the exam but didnt end up sitting for it. In Nigeria, Courses picked for waec and the performance on these courses determines what course the students will study in the university. Some students end up writing waec more than once because they failed some subjects needed for them to gain admission into the university. Some do the exam up to three times. During registration, you have to pick a specific number of subjects so they cant just register for let say two exact subject that they need. 

Generally, students are expected to write 9 subjects in WAEC and NECO. That's the maximum you can register and sit for. Also, you can write a minimum of 8 for the two exams. By this, it means, you can't register for subjects less than 8 (except for WAEC GCE that allows a minimum of 7 subjects). So what might be causing a spike in the major difference in the number of people that registered and number of people that sat for the exams is people are sitting for the exams they actually need and just registering some courses as miscellenous.

Total number of student that sat for waec in ondo state in 2006 were 37901 students. More tha half of the people that registered for the exam didnt actually sit for the exam. 

```python
#plotting subjects by number of people that registered for them
plt.figure(figsize=(10,5))
plt.bar(df2.SUBJECTS,df2.NUMBER_REGISTERED,color='red')
plt.xlabel("SUBJECTS")
plt.ylabel("NUMBER OF PEOPLE THAT REGISTERED")
plt.title("NUMBER OF PEOPLE THAT REGISTERED BY SUBJECTS")
plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.xticks(df2.SUBJECTS)
plt.show()
```

Courses that seem to have low registration rate are electives that people specializing in them take or they are just used as padding to  fulfill the 9 subjects requirement. Mathematics and english are compulsory courses for anyone writing waec so it comes as no suprise that a lot of people registered for it. Biology, Chemistry and Physics are core subjects for science students. After English, Yoruba is the second most registered language course and this is because Ondo state is a Yoruba land. CRK, Literature ande Government are core Art courses and Commerce and Accounting are core commercial courses. Geography and Economics are core courses taken across all departments.

```python
#plotting subject vs no of absent students
plt.figure(figsize=(10,5))
plt.bar(df2.SUBJECTS,df2.NUMBER_ABSENT,color='red')
plt.xlabel("SUBJECTS")
plt.ylabel("NUMBER OF PEOPLE THAT WERE ABSENT")
plt.title("NUMBER OF PEOPLE THAT ABSENT BY SUBJECTS")
plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.xticks(df2.SUBJECTS)
plt.show()
```

```python
#subject columns vs number of people that sat for waec
plt.figure(figsize=(10,5))
plt.bar(df2.SUBJECTS,df2.NUMBER_SAT,color='red')
plt.xlabel("SUBJECTS")
plt.ylabel("NUMBER OF PEOPLE THAT SAT")
plt.title("NUMBER OF PEOPLE THAT SAT BY SUBJECTS")
plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.xticks(df2.SUBJECTS)
plt.show()
```

Most people sat for biology and government. This might be because these are core subjects. And this can also be because they are subjects people are retaking the most.
